# Python Operators
An operator, in computer programing, is a symbol that usually represents an action or process.

## Bitwise Operators
<div>  </div>

```python
a = 2 # 0010
b = 3 # 0011
print(a & b)
#    0010
#    0011
# -> 0010 -> 2
```

```python
a = 2 # 
b = 3
print(bin(2))
print(bin(a)[2:].zfill(2))
print(bin(b)[2:])
bin(a | b)[2:].zfill(2)
```

```python
a = 1 # 
b = 2
print(bin(a)[2:].zfill(2))
print(bin(b)[2:])
bin(a & b)[2:].zfill(2)
```

```python
a = 5
```

```python
print(bin(a)[2:].zfill(4))
```

```python
print((a << 1))
```

```python
print((a >> 1))
```

```python
print(bin(2)[2:].zfill(8))
print(bin(2 << 2)[2:].zfill(8))
print(bin(2 >> 1)[2:].zfill(8))
```

# Variables

## Valid python variables

```python
# Name should not start with -> !@#^%@$@(*&)
# It must start with Alphabets Lower or upper case
# It should not start with number
```

```python
test = 1
test23 = 2
ABC = 2
aBc = 10
print(test)
print(test23)
print(ABC)
print(aBc)
```

```python
# @abc
# $test123
# 123ABC
```
