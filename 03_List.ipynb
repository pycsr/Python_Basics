# List

## - Definition

- It a `hetrogenious sequence data` types.
- A single list may contain DataTypes like Integers, Strings, as well as Objects. Lists are `mutable` data type and hence,they can be altered after their creation.
- Its an ordered data type

## Examples

```python
a = [] # empty list
print(a)
print(len(a))
print(type(a))
```

```python
a = [1, 2., 3.4, 2+3j, "PyCSR", [4,5,6], (1,2,3), {"A":3}]
print(a)
print(len(a))
```

## How to check type of a List

```python
print(type(a))
```

## Concatenation

```python
a = [1,2,3]
b = [4,5,6]
c = a+b
print(a)
print(b)
print(c)
print(len(c))
```

## Repetition

```python
a = [1,2,3]
c = a*5
print(a)
print(len(a))
print(c)
print(len(c))
```

## Indexing

<div>  </div>

```python
# Syntax: <variable_name>[<start>:<end>:<step_size>]
# start = by default 0/-1
# end = length-1/-lenght (end will always be excluded)
# Step size = +1 (-ve step to be used if backward slicing is used)
```

### Forward/positive Indexing

```python
a = ["P","Y","T","H","O","N"]
print(a[1])
print(a[3])
```

### Backward/negative Indexing

```python
a = ["P","Y","T","H","O","N"]
print(a[-1])
print(a[-3])
```

## - Slicing

<div>  </div>

### Forward Slicing

```python
a = ["P","Y","T","H","O","N"]
print(a[:])
print(a[::])
print(a[:4:])
print(a[:4:2])
print(a[2:])
print(a[2::3])
```

### Backward Slicing

```python
a = ["P","Y","T","H","O","N"]
print(a[-1:-5:])
print(a[-1:-5:-1])
print(a[-1::])
print(a[-1:4:-1])
print(a[-5:-2])
print(a[-2:-5:-1])
print(a[::-1])
```

## How to modify list elements

```python
a = [1,2,3,4,5]
print(a)
a[-1] = 50
print(a)
```

```python
a = [1,2,3,4,5]
print(a)
a[0:2] = [0,0,0]
print(a)
```

## How to delete a list

```python
a = [1,2,3,4,5]
print(a)
del a
print(a)
```
