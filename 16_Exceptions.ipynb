## Try - Except block

```python
# Syntax:
# try:
#     <Write your code which is doubtful>
# except:
#     <Other part of code or print execption info>
```

```python
a = 10
b = int(input("Enter a number.. "))
print(a/b)
print(a+b)
```

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except:
    print("Exception has occurred")
```


## Catching exception message

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except Exception as error: # Exception : mother class of all exceptions, error : is a variable name
    print("Exception occurred due to '{}'".format(error))
```

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except ZeroDivisionError as error: # Exception : mother class of all exceptions, error : is a variable name
    print("Exception occurred due to '{}'".format(error))
```


## Catching exception class name

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except ZeroDivisionError as error: # Exception : mother class of all exceptions, error : is a variable name
    print(error.__class__.__name__)
    print("Exception occurred due to '{}'".format(error))
```

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except Exception as error: # Exception : mother class of all exceptions, error : is a variable name
    print("Exception occurred due to '{}' which belongs to {} exception class".format(error,error.__class__.__name__))
```

## Try - Except and Finally block

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except ZeroDivisionError as error: # Exception : mother class of all exceptions, error : is a variable name
    print(error.__class__.__name__)
    print("Exception occurred due to '{}'".format(error))
finally:
    print("I am in finally block")
```

```python
a = 10
try:
    b = int(input("Enter a number.. "))
    print(a/b)
    print(a+b)
except ZeroDivisionError as error: # Exception : mother class of all exceptions, error : is a variable name
    print(error.__class__.__name__)
    print("Exception occurred due to '{}'".format(error))
finally:
    print("I am in finally block")
```
