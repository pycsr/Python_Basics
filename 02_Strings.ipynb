# String

## Definition

In Python, string is an `immutable` sequence data type. It is the sequence of Unicode characters wrapped inside `single`, `double`, or `triple` quotes.

## Various types for defining string data types

```python
a = 'This is first string'
b = "This is second string"
c = """This is third string"""
d = '''This is third string'''
e = 'This is "first" string'
f = "This is 'first' string"
g = "123ABC@#$%&@^#"
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
```

## Indexing

<div>  </div>

```python
# Syntax: <variable_nbame>[<index_number>]
```

```python
a = "0123456789"
print(len(a)) # --> len is a function to get number of items/chars/elements in given nstring
# print(a[100])
```

### Forward/positive Indexing

```python
a = "ABCDEFGHIJ"
print(len(a)) # --> len is a function to get number of items/chars/elements in given nstring
print(a[0])
print(a[1])
print(a[5])
```

### Backward/Negative Indexing

```python
a = "ABCDEFGHIJ"
print(len(a)) # --> len is a function to get number of items/chars/elements in given nstring
print(a[-1])
print(a[-2])
print(a[-5])
```

## Slicing

```python
# Syntax = <variable_name>[<start>:<end>:<step_size>]
# start -> default is 0
# end -> default length-1, if end index i sgiven then it will be excluded
# step_size -> +1
```

### Forward slicing

```python
a = "0123456789"
print(a)
print(a[::])
print(a[0::])
print(a[3::])
print(a[7::])
print(a[2:8:])
print(a[2:8:2])
print(a[2:8:3])
print(a[::2])
print(a[1::2])
```

### Backward slicing

<div>  </div>

```python
# step_size -> -1 must be mentioned as -ve
```

```python
a = "0123456789"
print(a)
# print(a[-1::])
print(a[-1:-5:-1])
print(a[-2::-2])
print(a[::-1])
```

# Various ways for string formatting

```python
a = "Pankaj"
b = "PyCSR"
print(a)
print(b)
```

```python
print(a + "is preparing " + b + "Video")
```

## - using {} braces

```python
print("{} is prepairing {} Video".format(a,b))
```

```python
print("{} is prepairing {} Video".format(b,a))
```

## - {} with indexing

```python
print("{0} is prepairing {1} Video".format(a,b))
```

```python
print("{1} is prepairing {0} Video".format(a,b))
```

## - using !r and !s flag

```python
print("{!r} is prepairing {!s} Video".format(a,b))
```

```python
print("{!r} is prepairing {} Video".format(a,b))
```

## - using left/right/center justification

```python
a = " PyCSR "
print(a)
```

```python
print("{:<20}".format(a))
```

```python
print("{:>20}".format(a))
```

```python
print("{:^20}".format(a))
```

## - justification with character padding

```python
print("{:#<20}".format(a))
```

```python
print("{:$>20}".format(a))
```

```python
print("{:@^20}".format(a))
```

## - Number formatting to binary, Octa, Hex

```python
a = 20
print("{:b}".format(a))
```

```python
a = 20
print("{:#b}".format(a))
```

```python
a = 20
print("{:#o}".format(a))
```

```python
a = 20
print("{:#x}".format(a))
```
