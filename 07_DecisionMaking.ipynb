# Python Decision Making

## `if statement`

```python
# Syntax:
# if <condition>:
#     <statements>
#     .
#     .
    
```

### Definition

Block of code will be executed iff given `condition is TRUE`

<div>  </div>

### Example

```python
a = [1,2,3,4,5]
if 20 in a:
    print("20 is present")
print("I am outside of if block")
```

## `if else statement`

### Definition

`else` block will be executed if `given condition in if block is FALSE`

<div>  </div>

### Example

```python
a = [1,2,3,4,5]
if 20 in a:
    print("20 is present")
else:
    print("20 is not present")
```

## `if.. elif ..else statement`

### Definition

- The `elif` is short for else if. It allows us to check for multiple expressions.

- If the condition for if is False, it checks the condition of the next elif block and so on.

- If all the conditions are False, the body of else is executed.

### Example

```python
a = [1,2,3,4,5]
if 20 in a:
    print("20 is present")
elif 10 in a:
    print("10 is present")
else:
    print("I am in else block")
```

```python
a = [1,2,3,4,5]
if 20 in a:
    print("20 is present")
elif 2 in a:
    print("2 is present")
else:
    print("I am in else block")
```

## `nested if else statement`

### Definition

Given if-else block of code will be executed based on whichever given `condition is TRUE`

### Example

```python
a = "Python"
if "P" in a:
    if "Py" in a:
        print("Py is present in {}".format(a))
    else:
        print("Py is not present in {}".format(a))
else:
    print("Nothing is present in {}".format(a))
        
    
```

```python
a = "Python"
if "Pi" in a:
    if "Py" in a:
        print("Py is present in {}".format(a))
    else:
        print("Py is not present in {}".format(a))
else:
    print("Nothing is present in {}".format(a))
        
    
```

```python
a = "Python"
if "P" in a:
    if "Pi" in a:
        print("Pi is present in {}".format(a))
    else:
        print("Pi is not present in {}".format(a))
else:
    print("Nothing is present in {}".format(a))
        
    
```

## Using if..else along with for loop

```python
a = [1,2,3,4,5]
for elm in a:
    print(elm)
```

```python
a = [1,2,3,4,5]
for elm in a:
    if elm%2 == 0:
        print("Even number is {}".format(elm))
```

```python
a = [1,2,3,4,5]
for elm in a:
    if elm%2 == 0:
        print("Even number is {}".format(elm))
    else:
        print("Odd number is {}".format(elm))
    
```
