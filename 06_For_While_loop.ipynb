# Python Control Structures

## `For loop`

### Syntax

```python
# for <variable_name> in <any_sequence/Iterator>:
#     <statements>
#     .
#     .    
```

### Definition

- To be used when user knows how many iteration to be executed
- For loop to be executed to execute block of code several times

<div>  </div>

### Example

```python
a = [1,2,3,4,5]
print(a)
```

```python
for var in a:
    print(var)
```

```python
a = "Hello World!!"
for char in a:
    print(char)
```

## `While loop`

### Syntax

```python
# while <condition>:
#     <statements>
#     .
#     .
```

### Definition

- To be used when user is not aware of how many iteration will be executed

`or`
- While loop keeps executing until given condition becomes FALSE

<div>  </div>

### Example

```python
cnt = 10
while cnt >= 0:
    print(cnt)
    cnt = cnt-1 # cnt -= 1
```

```python
cnt = 0
while cnt <= 20:
    print(cnt)
    cnt = cnt+1 # cnt += 1
```

## `break statement`

<div>  </div> 

### Definition

- The break statement is used for premature termination of the current loop.

- break statement can be used in both while and for loops.

- If you are using nested loops, the break statement stops the execution of the `innermost loop` and starts executing the next line of the code after the block.

### Example

```python
word = "Python"
for char in word:
    print(char)
```

```python
word = "Python"
for char in word:
    print(char)
    if char == "h":
        break
```

## `continue statement`

<div>  </div> 

### Definition

- The continue statement in Python returns the control to the beginning of the current loop. When encountered, the `loop starts next iteration without executing the remaining statements` in the current iteration.

- The continue statement can be used in both while and for loops.

### Example

```python
word = "Python"
for char in word:
    if char == "h":
        continue
    else:
        print(char)
```

```python
word = "Python"
for char in word:
    if char == "h":
        continue
        print("Ok, I am in continue block") # This will never be executed
    else:
        print(char)
```

## `pass statement`

### Definition

- It is used when a `statement is required syntactically` but you `do not want any command or code to execute`.

- The pass statement is a null operation; nothing happens when it executes. The pass statement is also useful in places where your code will eventually go, but has not been written yet.

### Example

```python
a = [1,2,3,4,5]
for i in a:
    pass
    print(i)
```
